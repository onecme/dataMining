```python
# Menghapus data yang terdeteksi duplikat pada dataframe
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

```python
df['Items'].unique()
```

```python
# Mengubah data ke dalam format one-hot encoding
df_pivot= df.pivot_table(index='TransactionNo', columns='Items', aggfunc='size', fill_value=0)
df_pivot
```

```python
# Mengubah ke bentuk binary (1 jika item dibeli, 0 jika tidak)
df_binary = df_pivot.applymap(lambda x: 1 if x > 0 else 0)
df_binary
```

```python
# Mencari itemsets yang sering muncul menggunakan algoritma apriori
frequent_itemsets = apriori(df_binary, min_support=0.05, use_colnames=True)
frequent_itemsets
```

```python
# Menarapkan aturan asosiasi dan menampilkan hasil berdasarkan lift
rules = association_rules(frequent_itemsets, metric='lift', min_threshold=1.0)
rules
```
